```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import sys
import csv
import pandas as pd
import time
```

```python
def scrape_data(url, path):
    """ This function is for scraping data from the google playstore. This is for three
    applications i.e. Pandora Music, Spotify and Amazon Music. Selenium Package and Beautiful soup are used 
    for web scrapping. This function takes url and path as parameters.
    :param url: http link
    :param path: computer location
    >>> x= scrape_data(url, path)
    >>> s = x['Approval'][0] > 0"""
    driver = webdriver.Chrome("C:/Users/srikanth/PycharmProjects/PR Project/chromedriver_win32 (1)/chromedriver.exe")
    driver.get(url)
    time.sleep(1)

    html2=''
    count=0
    html2 = driver.execute_script("return document.body.innerHTML;")

    while count<10000:
        
        try:
            el3 = driver.find_element_by_xpath("//*[@class='RveJvd snByac']")
            el3.click()  
            html2 = driver.execute_script("return document.body.innerHTML;")
        except:
            print('scroll error')
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            count+=1

    html2 = driver.execute_script("return document.body.innerHTML;")
    driver.close()
    soup = bs(html2,'html.parser')
    driver.quit()
    id=1
    idlist=[]
    usernamelist=[]
    starcountlist=[]
    aprrovallist=[]
    commentlist=[]
    link=''
    df=pd.DataFrame()
    counter=1


    for i,tag in enumerate(soup.find_all(['div'],attrs={'jsmodel':['y8Aajc']})):
        usernameClass=tag.find('span',attrs={'class':'X43Kjb'})
        starcount=tag.find_all('div',attrs={'class':'vQHuPe bUWb7c'})
        starcount=len(starcount)
        approval=tag.find('div',attrs={'class':'jUL89d y92BAb'}).text
        username=usernameClass.text
        textComment=tag.find('span',attrs={'jsname':'bN97Pc'}).text
        textCommentExtra=tag.find('span',attrs={'jsname':'fbQN7e'}).text.strip()
        if(textCommentExtra!=''):
            textComment=textCommentExtra
            idlist.append(id)
            usernamelist.append(username)
            starcountlist.append(starcount)
            aprrovallist.append(approval)
            commentlist.append(textComment)

    df['Id']=idlist
    df['Username']=usernamelist
    df['Stars']=starcountlist
    df['Approval']=aprrovallist
    df['Review']=commentlist
    #print(df)
    df.to_csv(path, encoding='utf-8')

    return df
```
